### Вопрос 2, 3
Постройте графики разных функции с помощью matplotlib на одном графике, сделайте подписи осей, добавте легенду, сохраните график в файл. 

```python
X = np.linspace(-5,5, 50)
y = [np.sin(x) for x in X]
plt.plot(X,y, label = 'sin')

y = [np.cos(x) for x in X]
plt.plot(X,y, label = 'cos')

y = [x*0.5 + 2 for x in X]
plt.plot(X,y, label = 'line')

plt.legend()
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title("Графики функций")

plt.savefig('function_plots.png')
```

Постройте на одном изображении 4 графика: параболу, гиперболу, экспоненту и логарифм

```python
plt.rcParams['figure.figsize'] = 10, 7
X = np.linspace(-10,10, 20)
funcArray = [lambda x: x ** 2, lambda x: x**3, lambda x: np.exp(x), lambda x: np.log(x)]
funcNames = ['$x^2$', '$x^3$', 'exp', 'log']

for i,f,n in zip([221, 222, 223, 224],funcArray, funcNames):
    plt.subplot(i)
    y = [f(x) for x in X]
    plt.plot(x, y, 'r')
    plt.title(n, fontsize = 10)
```

### Вопрос 5
Постройте heatmap для заданной функции

```python
def myFunction(x,y):
    return 3*x*y + x - 2*y

x = []
y = []
f = []

for i in range(0,5,1):
    for j in range(0,5,1):
        x.append(i)
        y.append(j)
        f.append(myFunction(i,j))


myData = pd.DataFrame()
myData['x'] = x
myData['y'] = y
myData['f'] = f

myData.head()

result = myData.pivot(index='x', columns='y', values='f')
sns.heatmap(result, annot=True, fmt="g", cmap='viridis')
```
