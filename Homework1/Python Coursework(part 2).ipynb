### Вопрос 1

Скачать с новостного сайта текст статьи без вкраплений JS и HTML-тегов

```python
import re
import urllib
from bs4 import BeautifulSoup
 
html = urllib.request.urlopen('http://time.com/5468842/united-states-china-trade-war-cease-fire/')
soup = BeautifulSoup(html, "lxml")
data = soup.findAll(text=True)
result = filter(visible, data)
 
articleText = list(result)
```

```python
def visible(element):
    if element.parent.name in ['style', 'script', '[document]', 'head', 'title']:
        return False
    elif re.match('<!--.*-->', str(element.encode('utf-8'))):
        return False
    return True
```

```python
#print out article text
for i in articleText:
    if i != '\n':
        print (i)
```
